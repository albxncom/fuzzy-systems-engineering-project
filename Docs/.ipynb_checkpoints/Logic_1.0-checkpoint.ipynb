```python
import numpy as np
import pandas as pd
import json


DEBUG=True

# the data frame
df = pd.read_csv('./../data.csv')

# define here lingustic terms and how to fuzzify
crispify_binary = lambda x: 1 if x == 'yes' else 0
fuzzify_goodness = lambda x: general_fuzzifier(x, ["very poor", "poor", "moderate", "good", "very good"])
fuzzify_agreement = lambda x: general_fuzzifier(x, ["strongly disagree", "disagree", "neither agree nor disagree", "agree", "strongly agree"])
fuzzify_quantity = lambda x: general_fuzzifier(x, ["very few", "a few", "moderate", "some", "very much"])
fuzzify_difficulty = lambda x: general_fuzzifier(x, ["very difficult", "difficult", "moderate", "easy", "very easy"])
fuzzify_difficulty = lambda x: general_fuzzifier(x, ["very difficult", "difficult", "moderate", "easy", "very easy"])
fuzzify_satisfaction = lambda x: general_fuzzifier(x, ["very satisfied", "satisfied", "neither", "dissatisfied", "very dissatisfied"])

def general_fuzzifier(x, labels):
    fuzzifzied = {}
    fuzzifzied.setdefault(labels[0], np.array([0, 0, 0.5, 1]))
    fuzzifzied.setdefault(labels[1], np.array([0.5, 1, 1.5, 2]))
    fuzzifzied.setdefault(labels[2], np.array([1.5, 2, 2.5, 3]))
    fuzzifzied.setdefault(labels[3], np.array([2.5, 3, 3.5, 4]))
    fuzzifzied.setdefault(labels[4], np.array([3.5, 4, 4.5, 4.5]))
    return fuzzifzied[x]


#some cleanup
df.iloc[:,2] = df.iloc[:,2].apply(lambda x: x.split(",")) #course_category
df.iloc[:,4] = df.iloc[:,4].apply(crispify_binary) #exercices_mandatory

#fuzzify the data
df.iloc[:,df.columns.get_loc('language_professor')] = df.iloc[:,df.columns.get_loc('language_professor')].apply(fuzzify_goodness) #language_professor
df.iloc[:,df.columns.get_loc('quality_language_lectures')] = df.iloc[:,df.columns.get_loc('quality_language_lectures')].apply(fuzzify_goodness) #quality_language_lectures
df.iloc[:,df.columns.get_loc('language_assistant')] = df.iloc[:,df.columns.get_loc('language_assistant')].apply(fuzzify_goodness) #language_assistant
df.iloc[:,df.columns.get_loc('quality_language_exercises')] = df.iloc[:,df.columns.get_loc('quality_language_exercises')].apply(fuzzify_goodness) #quality_language_exercises
df.iloc[:,df.columns.get_loc('content_is_understandable')] = df.iloc[:,df.columns.get_loc('content_is_understandable')].apply(fuzzify_agreement) #content_is_understandable
df.iloc[:,df.columns.get_loc('examples')] = df.iloc[:,df.columns.get_loc('examples')].apply(fuzzify_agreement) #examples
df.iloc[:,df.columns.get_loc('difficulty_lectures')] = df.iloc[:,df.columns.get_loc('difficulty_lectures')].apply(fuzzify_difficulty) #difficulty_lectures
df.iloc[:,df.columns.get_loc('difficulty_exercices')] = df.iloc[:,df.columns.get_loc('difficulty_exercices')].apply(fuzzify_difficulty) #difficulty_exercices
df.iloc[:,df.columns.get_loc('listening_professor')] = df.iloc[:,df.columns.get_loc('listening_professor')].apply(fuzzify_goodness) #listening_professor
df.iloc[:,df.columns.get_loc('responsiveness_professor')] = df.iloc[:,df.columns.get_loc('responsiveness_professor')].apply(fuzzify_goodness) #responsiveness_professor
df.iloc[:,df.columns.get_loc('responsiveness_assistant')] = df.iloc[:,df.columns.get_loc('responsiveness_assistant')].apply(fuzzify_goodness) #responsiveness_assistant
df.iloc[:,df.columns.get_loc('course_structure')] = df.iloc[:,df.columns.get_loc('course_structure')].apply(fuzzify_agreement) #course_structure
df.iloc[:,df.columns.get_loc('course_independently')] = df.iloc[:,df.columns.get_loc('course_independently')].apply(fuzzify_agreement) #course_independently
df.iloc[:,df.columns.get_loc('course_evaluation_modalities')] = df.iloc[:,df.columns.get_loc('course_evaluation_modalities')].apply(fuzzify_agreement) #course_evaluation_modalities
df.iloc[:,df.columns.get_loc('course_activities')] = df.iloc[:,df.columns.get_loc('course_activities')].apply(fuzzify_agreement) #course_activities
df.iloc[:,df.columns.get_loc('expectations_lectures')] = df.iloc[:,df.columns.get_loc('expectations_lectures')].apply(fuzzify_agreement) #expectations_lectures
df.iloc[:,df.columns.get_loc('course_prepared_exam')] = df.iloc[:,df.columns.get_loc('course_prepared_exam')].apply(fuzzify_agreement) #course_prepared_exam
df.iloc[:,df.columns.get_loc('effort_lectures')] = df.iloc[:,df.columns.get_loc('effort_lectures')].apply(fuzzify_quantity) #effort_lectures
df.iloc[:,df.columns.get_loc('effort_exercises')] = df.iloc[:,df.columns.get_loc('effort_exercises')].apply(fuzzify_quantity) #effort_exercises
df.iloc[:,df.columns.get_loc('effort_exam_preperation')] = df.iloc[:,df.columns.get_loc('effort_exam_preperation')].apply(fuzzify_quantity) #effort_exam_preperation
df.iloc[:,df.columns.get_loc('exam_difficulty')] = df.iloc[:,df.columns.get_loc('exam_difficulty')].apply(fuzzify_difficulty) #exam_difficulty
df.iloc[:,df.columns.get_loc('open_book_exam')] = df.iloc[:,df.columns.get_loc('open_book_exam')].apply(crispify_binary) #open_book_exam
df.iloc[:,df.columns.get_loc('grading')] #grading
df.iloc[:,df.columns.get_loc('statisfaction')] = df.iloc[:,df.columns.get_loc('statisfaction')].apply(fuzzify_satisfaction) #statisfaction
```

```python
#compute mean for the following labels
means = {}
labels = ["language_professor","quality_language_lectures","language_assistant","quality_language_exercises","content_is_understandable","examples","difficulty_lectures","difficulty_exercices","listening_professor","responsiveness_professor","responsiveness_assistant","course_structure","course_independently","course_evaluation_modalities","course_activities","expectations_lectures","course_prepared_exam","effort_lectures","effort_exercises","effort_exam_preperation","exam_difficulty","grading","statisfaction"]
for label in labels:
    means.setdefault(label, df.groupby(by='course_id')[label].mean())

means = pd.DataFrame(means)
dropped = df.drop(columns=labels).drop_duplicates(subset=['course_id'])

means = dropped.merge(means, how='outer', on='course_id')
means
```

```python
DEBUG=False
def aggregate(l_, frame, weights=None, normalize=True):
    if DEBUG: print("AGGREGATING:",l_,weights)
    l = [ frame[col] for col in l_]
    
    if weights == None: 
        weights = np.array([1/len(l_)]*len(l_))
    elif normalize == True:
        weights = [ w/sum(weights) for w in weights ]

    if DEBUG: print("ADJ WEIGHTS:",weights)
    
    return sum([ w*f for (w,f) in zip(weights,l)])
```

```python
# aggregate certain questions
frame = means

#q_6 = language_lecture
language_lecture = ['language_professor', 'quality_language_lectures']
frame['language_lecture'] = aggregate(language_lecture, frame)

#q_7 = language_exercises
language_exercises = ['language_assistant', 'quality_language_exercises']
frame['language_exercises'] = aggregate(language_exercises, frame)

#q_8 = understandability
understandability = ['content_is_understandable', 'examples']
frame['understandability'] = aggregate(understandability, frame)
'language_lecture' : ['language_professor', 'quality_language_lectures']
'language_exercises' : ['language_assistant', 'quality_language_exercises']
'understandability' : ['content_is_understandable', 'examples']
'difficulty' : ['difficulty_lectures', 'difficulty_exercices']
'commitment' : ['listening_professor','responsiveness_professor','responsiveness_assistant']
'quality' : ['course_structure','course_independently','course_evaluation_modalities', 'course_activities', 'expectations_lectures', 'course_prepared_exam']
'effort' : ['effort_lectures','effort_exercises','effort_exam_preperation']
```

```python



import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz


# Generate trapezoidal membership function on range [0, 1]
x = np.arange(0, 5.05, 0.1)
mfx = fuzz.trapmf(x, [2, 2.5, 3, 4.5])

# Defuzzify this membership function five ways
defuzz_centroid = fuzz.defuzz(x, mfx, 'centroid')  # Same as skfuzzy.centroid
defuzz_bisector = fuzz.defuzz(x, mfx, 'bisector')
defuzz_mom = fuzz.defuzz(x, mfx, 'mom')
defuzz_som = fuzz.defuzz(x, mfx, 'som')
defuzz_lom = fuzz.defuzz(x, mfx, 'lom')

# Collect info for vertical lines
labels = ['centroid', 'bisector', 'mean of maximum', 'min of maximum',
          'max of maximum']
xvals = [defuzz_centroid,
         defuzz_bisector,
         defuzz_mom,
         defuzz_som,
         defuzz_lom]
colors = ['r', 'b', 'g', 'c', 'm']
ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]

# Display and compare defuzzification results against membership function
plt.figure(figsize=(8, 5))

plt.plot(x, mfx, 'k')
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)
plt.ylabel('Fuzzy membership')
plt.xlabel('Universe variable (arb)')
plt.ylim(-0.1, 1.1)
plt.legend(loc=2)

plt.show()
```
