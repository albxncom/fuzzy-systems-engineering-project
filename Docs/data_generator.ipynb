# Data Generator

This notebook can be used to generate random data.

```python
import random
import numpy as np

courses = [["User Centered Design","3018"],["Seminar Law and Computer Science: Distributed Trust in Finance","1892"],["Advanced Software Engineering","13000"],["Cloud Computing","12083"],["Seminar Software Composition","21875"],["Concurrency: Multi-core Programming and Data Processing","22027"],["Digital Humanities","32108"],["Computer Vision","31062"],["Automata on Infinite Structures","43024"],["Seminar Cryptography and Data Security","11867"],["Fuzzy Systems Engineering","53105"],["Emerging Technologies in Digitalisation","53891"],["Statistical Learning Methods with R","32079"],["Social Media Analytics","63091"]]
courses = [["User Centered Design","3018","t0,t1"],["Seminar Law and Computer Science: Distributed Trust in Finance","1892","t0"],["Advanced Software Engineering","13000","t1"],["Cloud Computing","12083","t1"],["Seminar Software Composition","21875","t1,t2"],["Concurrency: Multi-core Programming and Data Processing","22027","t2"],["Digital Humanities","32108","t3"],["Computer Vision","31062","t3"],["Automata on Infinite Structures","43024","t4"],["Seminar Cryptography and Data Security","11867","t4,t5"],["Fuzzy Systems Engineering","53105","t5"],["Emerging Technologies in Digitalisation","53891","t5"],["Statistical Learning Methods with R","32079","t6"],["Social Media Analytics","63091","t6"]]
           
#linguistic answers
l_tracks = ["t0", "t1", "t2", "t3", "t4", "t5", "t6"]
l_language = ["german", "french", "english"]
l_goodness = ["very poor", "poor", "moderate", "good", "very good"]
l_agreement = ["strongly disagree", "disagree", "neither agree nor disagree", "agree", "strongly agree"]
l_quantity = ["very few", "a few", "moderate", "some", "very much"]
l_difficulty = ["very difficult", "difficult", "moderate", "easy", "very easy"]
l_binary = ["yes", "no"]
l_difficulty = ["very difficult", "difficult", "moderate", "easy", "very easy"]
l_satisfaction = ["very satisfied", "satisfied", "neither", "dissatisfied", "very dissatisfied"]
l_grades = np.array([ i for i in range(100, 625, 25)]) / 100

data = []
data.append(["course_name","course_id","course_category","course_language","exercices_mandatory","language_professor","quality_language_lectures","language_assistant","quality_language_exercises","content_is_understandable","examples","difficulty_lectures","difficulty_exercices","listening_professor","responsiveness_professor","responsiveness_assistant","course_structure","course_independently","course_evaluation_modalities","course_activities","expectations_lectures","course_prepared_exam","effort_lectures","effort_exercises","effort_exam_preperation","exam_difficulty","open_book_exam","grading","statisfaction"]) #header

#for every course
for course in courses:
    
    number_of_samples = random.randint(5,10) #do between 5-10 samples
    
    fixed_lang = l_language[random.randint(0,len(l_language)-1)]
    fixed_exercises = l_binary[random.randint(0,len(l_binary)-1)]
    fixed_openbook = l_binary[random.randint(0,len(l_binary)-1)]
    
    for i in range(0,number_of_samples):
        q_1 = course[0]
        q_2 = course[1]
        q_3 = course[2]
        q_4 = fixed_lang
        q_5 = fixed_exercises
        q_6_1 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_6_2 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_7_1 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_7_2 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_8_1 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_8_2 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_9_1 = l_difficulty[random.randint(0,len(l_difficulty)-1)]
        q_9_2 = l_difficulty[random.randint(0,len(l_difficulty)-1)]
        q_10_1 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_10_2 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_10_3 = l_goodness[random.randint(0,len(l_goodness)-1)]
        q_11_1 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_11_2 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_11_3 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_11_4 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_11_5 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_11_6 = l_agreement[random.randint(0,len(l_agreement)-1)]
        q_12_1 = l_quantity[random.randint(0,len(l_quantity)-1)]
        q_12_2 = l_quantity[random.randint(0,len(l_quantity)-1)]
        q_12_3 = l_quantity[random.randint(0,len(l_quantity)-1)]
        q_13 = l_difficulty[random.randint(0,len(l_difficulty)-1)]
        q_14 = fixed_openbook
        q_15 = l_grades[random.randint(0,len(l_grades)-1)]
        q_16 = l_satisfaction[random.randint(0,len(l_satisfaction)-1)]
        
        q = [q_1,q_2,q_3,q_4,q_5,q_6_1,q_6_2,q_7_1,q_7_2,q_8_1,q_8_2,q_9_1,q_9_2,q_10_1,q_10_2,q_10_3,q_11_1,q_11_2,q_11_3,q_11_4,q_11_5,q_11_6,q_12_1,q_12_2,q_12_3,q_13,q_14,q_15,q_16]
        data.append(q)

        
```

## Save the data

```python
import pandas as pd 
df = pd.DataFrame(data)
header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = header #set the header row as the df header
df.to_csv("./../data.csv", index=None)
df
```
